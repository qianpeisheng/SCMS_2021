```python
# construct history
history_length = 5
_df = df[['main_load', 'rated_load_capacity', 'amplitude', 'angle']]
temp_list = []
for i in range(3600):
    data = _df.iloc[i:i+history_length,:]
    data = data.to_numpy().flatten()
    temp_list.append(data)
```

```python
X = np.array(temp_list)
X_train = X[:3000]
X_test = X[3000:]
Y = np.random.randint(2, size=3600)
Y_train = Y[:3000]
Y_test = Y[3000:]
```
